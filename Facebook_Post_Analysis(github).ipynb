```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Prepare the engagement data
engagements = startType_interaction[['num_comments', 'num_reactions', 'num_shares', 'total_interactions']]
engagements.index = ['Link', 'Photo', 'Status', 'Video']

# Clean column names for display
engagements.columns = (
    engagements.columns
    .str.replace('num_', '', regex=False)
    .str.replace('_', ' ', regex=False)
)

# Compute normalization parameters
threshold = 200000  # or any value you want to visually center around
vmin = engagements.min().min()
vmax = engagements.max().max()
divnorm = mcolors.TwoSlopeNorm(vmin=vmin, vcenter=threshold, vmax=vmax)

# Create a custom diverging colormap
cmap_lower = plt.cm.Blues_r
cmap_upper = plt.cm.Oranges
colors = np.vstack((
    cmap_lower(np.linspace(0, 1, 128)),
    cmap_upper(np.linspace(0, 1, 128))
))
combined_cmap = mcolors.LinearSegmentedColormap.from_list('CustomSplitMap', colors)

# Set figure size dynamically based on shape
cell_width = 1.25
cell_height = 1.25
n_rows, n_cols = engagements.shape
fig_width = n_cols * cell_width
fig_height = n_rows * cell_height

# Plot the heatmap with custom normalization and colormap
plt.figure(figsize=(fig_width * 1.5, fig_height * 1.2))
ax = sns.heatmap(
    engagements,
    annot=True,
    fmt=' ',
    linewidths=.5,
    cmap=combined_cmap,
    norm=divnorm,
    square=True
)

for text in ax.texts:
    val = float(text.get_text())
    text.set_text(f'{val:,.0f}')

# Labels and layout
plt.title('Interactions by Type of Post', pad=20)
ax.tick_params(axis='both', length=0)
ax.set(xlabel='Interaction Types', ylabel='Post Type')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Reset index
pivot_df = date_time_interaction2.reset_index()

# Convert publish_hour to integer (0–23)
pivot_df['publish_hour'] = pivot_df['publish_hour'].apply(lambda x: x.hour)

# Get all status types
status_types = pivot_df['status_type'].unique()

# Define the desired order for days of the week
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 13), sharex=False, sharey=True)
axs = axs.flatten()

# Loop over each status_type
for i, status in enumerate(status_types):
    ax = axs[i]

    # Filter and sort data by hour
    data = pivot_df[pivot_df['status_type'] == status].sort_values('publish_hour')

    # Plot each day in the correct order if it exists in the data
    for day in ordered_days:
        if day in data.columns:
            ax.plot(data['publish_hour'], data[day], label=day)

    # Title and labels
    ax.set_title(f'{status.capitalize()} Posts by Hour')
    ax.set_xlabel('Hour of Day')
    ax.set_ylabel('Count')
    ax.grid(True)

    # Sort and display legend by day order
    handles, labels = ax.get_legend_handles_labels()
    label_order = [day for day in ordered_days if day in labels]
    handles_ordered = [handles[labels.index(day)] for day in label_order]
    ax.legend(handles_ordered, label_order, title='Day')

# Set x-ticks for all subplots
for ax in axs:
    ax.set_xticks(range(0, 24))
    ax.set_xticklabels(range(0, 24))
    ax.set_xlabel('Hour of Day')

# Adjust layout
plt.tight_layout(pad=3.0)
plt.show()
```

```python
type_reaction = pd.pivot_table(
    df2,
    index='status_type',
    values=['num_likes',
            'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys'],
    aggfunc='sum',
    fill_value=0
)

type_reaction['total_posts'] = df2['status_type'].value_counts()

type_reaction['total_reactions'] = type_reaction.sum(axis=1)
type_reaction['positive_sentiment'] = (
    type_reaction['num_likes'] +
    type_reaction['num_loves'] +
    type_reaction['num_wows'] +
    type_reaction['num_hahas']
)

type_reaction['negative_sentiment'] = (
    type_reaction['num_sads'] +
    type_reaction['num_angrys']
)

percentage_columns = (type_reaction.iloc[:, :-1].div(type_reaction['total_reactions'], axis=0) * 100).round(3)

percentage_columns = percentage_columns.add_suffix('_pct')

type_reaction = pd.concat([type_reaction, percentage_columns], axis=1)


type_reaction['positive_sentiment_pct'] = (type_reaction['positive_sentiment'] / (type_reaction['total_reactions']) * 100).round(3)
type_reaction['negative_sentiment_pct'] = (type_reaction['negative_sentiment'] / (type_reaction['total_reactions']) * 100).round(3)
type_reaction
```

```python
type_reaction2 = type_reaction.copy()
means = type_reaction2.mean(axis=1)
type_reaction2[''] = np.nan
type_reaction2['total'] = type_reaction2['total_reactions'] / type_reaction2['total_posts']

type_reaction2.loc[''] = np.nan
column_sum = type_reaction.mean(axis=0)
type_reaction2.loc['average'] = column_sum
type_reaction2
```

```python
import matplotlib.colors as mcolors
# Calculate average reactions per post
reaction = ['num_angrys', 'num_hahas', 'num_likes', 'num_loves', 'num_sads', 'num_wows', '', 'total']
avg_reaction2 = pd.DataFrame()
for re in reaction:
    if re =='':
        avg_reaction2[''] = type_reaction2[re]
    elif re == 'total':
        avg_reaction2['all reactions'] = type_reaction2[re]
    else:
        name = re.split('_')[1]
        avg_reaction2[name] = type_reaction2[re] / type_reaction2['total_posts']

# Parameters
threshold = 2.0
vmin = avg_reaction2.min().min()
vmax = avg_reaction2.max().max()

# Create a TwoSlopeNorm normalization centered at threshold
divnorm = mcolors.TwoSlopeNorm(vmin=vmin, vcenter=threshold, vmax=vmax)

# Create two colormaps: one for below threshold, one for above
cmap_lower = plt.cm.Blues_r  # reversed Blues for lower range
cmap_upper = plt.cm.Oranges  # Oranges for upper range

# Sample colors from each colormap
colors_lower = cmap_lower(np.linspace(0, 1, 128))
colors_upper = cmap_upper(np.linspace(0, 1, 128))

# Stack colors vertically to combine
colors = np.vstack((colors_lower, colors_upper))

# Create combined colormap
combined_cmap = mcolors.LinearSegmentedColormap.from_list('TwoPartCmap', colors)

# Set figure size
cell_width = 1.25
cell_height = 1.25
n_rows, n_cols = avg_reaction2.shape
fig_width = n_cols * cell_width
fig_height = n_rows * cell_height

# Plot heatmap
plt.figure(figsize=(fig_width * 1.5, fig_height * 1.2))
ax = sns.heatmap(
    avg_reaction2,
    annot=True,
    fmt='.4f',
    linewidths=.5,
    cmap=combined_cmap,
    norm=divnorm,
)

plt.title('Average Reactions per Post Type heatmap', pad=20)
ax.tick_params(axis='both', length=0)
plt.xlabel('Reaction')
plt.ylabel('Post Type')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
